```python
def give_entire_relative_abun(data,string_name):

  data_1 = data[data["dataset2.Host_disease"] != "Adenosquamous carcinoma"]
  unique_taxa = []
  temp_list = np.array(data_1[string_name])

  for i in range(len(temp_list)):
    if (temp_list[i] not in unique_taxa):
      unique_taxa.append(temp_list[i])

  #Removing na
  unique_taxa = pd.DataFrame(unique_taxa).dropna()

  #Flattening the array
  unique_taxa = np.array(unique_taxa).ravel()

  print("Number of Unique Taxa are:-",len(unique_taxa))

  counts_per_taxa = []

  total_counts = sum(data_1["count"])

  for i in range(len(unique_taxa)):
    data_temp = data_1[data_1[string_name]==unique_taxa[i]]

    temp_counts_1 = sum(data_temp["count"])*100/total_count

    counts_per_taxa.append(temp_counts_1)

  temp_data = pd.DataFrame()
  temp_data[string_name] = unique_taxa
  temp_data["relative_abundance"] =  counts_per_taxa

  return temp_data


```

```python
def give_class_relative_abun(data,string_name):

  #Calculating the class wise independently
  ac_data = data[data["dataset2.Host_disease"]=="Adenocarcinoma"]
  sc_data = data[data["dataset2.Host_disease"]=="Squamous cell carcinoma"]


  total_ac = sum(ac_data["count"])
  total_sc = sum(sc_data["count"])

  counts_per_taxa_ac = []
  counts_per_taxa_sc = []

  unique_taxa = []
  temp_list = np.array(data[string_name])

  for i in range(len(temp_list)):
    if (temp_list[i] not in unique_taxa):
      unique_taxa.append(temp_list[i])

  #Removing na
  unique_taxa = pd.DataFrame(unique_taxa).dropna()

  #Flattening the array
  unique_taxa = np.array(unique_taxa).ravel()

  print("Number of Unique Taxa are:-",len(unique_taxa))

  for i in range(len(unique_taxa)):
    ac_data_temp = ac_data[ac_data[string_name]==unique_taxa[i]]
    temp_counts_ac = sum(ac_data_temp["count"])*100/total_ac
    counts_per_taxa_ac.append(temp_counts_ac)

    sc_data_temp = sc_data[sc_data[string_name]==unique_taxa[i]]
    temp_counts_sc = sum(sc_data_temp["count"])*100/total_sc
    counts_per_taxa_sc.append(temp_counts_sc)

  temp_data = pd.DataFrame()
  temp_data[string_name] = unique_taxa

  temp_data["AC"] = counts_per_taxa_ac
  temp_data["SC"] = counts_per_taxa_sc



  return temp_data

```

***Finding Accuracy, Precision, Specificity and Sensitivity***

```python
def find_A_P_SP_SN(data):
  data_np = np.array(data)
  A = []
  P = []
  SP = []
  SN = []
  F1 = []

  for i in range(len(data_np)):
    A.append(sum(data_np[i,[1,4]])/sum(data_np[i,1:]))
    P.append(data_np[i,4]/sum(data_np[i,[3,4]]))
    SP.append(data_np[i,1]/sum(data_np[i,[1,3]]))
    SN.append(data_np[i,4]/sum(data_np[i,[2,4]]))
    F1.append(data_np[i,4]/(data_np[i,4]+((data_np[i,2]+data_np[i,3])/2)))

  new_data = copy.deepcopy(data)

  new_data["Accuracy"] = np.array(A).round(2)
  new_data["Precision"] = np.array(P).round(2)
  new_data["Specificity"] = np.array(SP).round(2)
  #new_data["Sensitivity"] = np.array(SN).round(2)
  new_data["F1 Score"] = np.array(F1).round(2)

  return new_data
result_data_2 = find_A_P_SP_SN(result_data_non_lda)
result_data_2["AUC"] = auc_1
result_data_2.to_csv("Result_Data_2.csv")

results_data_1 = find_A_P_SP_SN(results_data)
results_data_1["AUC"] = auc_2
results_data_1.to_csv("Result_Data_1.csv")
```
